**6. Handling Missing Values**

```python
#No. of Null Values for Each Col
df.isnull().sum()
```

```python
#Replacing Numeric nulls with mean:
for col in df.columns:
  if(df[col].dtype=="float64"):
    col_mean=df[col].mean()
    df[col]=df[col].fillna(col_mean)

df.isnull().sum()
```

```python
#Replacing categorical with string

for col in df.columns:
  if(df[col].dtype=="object"):
    col_mode=df[col].mode()[0]
    df[col]=df[col].fillna(col_mode)

df.isnull().sum()
```

**7. Type Conversion**

```python
#Converting object to string
for col in df.columns:
  if df[col].dtype=="object":
    df[col]=df[col].astype("string")

df.dtypes
```

```python
#Replacing string with float format

df["Height"]=df["Height"].astype('string')
df["Height"]=df["Height"].str.replace("-",".",regex=False)

df["Height"]=df["Height"].astype("float")
df["Height"]
```

```python
df['Salary']=df['Salary'].astype("float")
df['Salary']=df['Salary'].astype("int")

df.dtypes
```
