```python
for col in df.columns:
    if df[col].dtype=='object':
      df[col]=df[col].fillna(df[col].mode()[0])
```

```python
for col in df.columns:
  if df[col].dtype=="float":
    mean=df[col].mean()
    df[col]=df[col].fillna(mean)

for col in df.columns:
  if df[col].dtype=="float":
    df[col]=df[col].astype("int")

df.isnull().sum()
```

```python
df["Height"]=df["Height"].astype("string")
df["Height"]=df["Height"].str.replace("-", ".", regex=False)
df["Height"]=df["Height"].astype("float")

df.dtypes
```

```python
mean={}
for col in df.columns:
    if df[col].dtype in ['int','float64']:
        total=0
        count=0
        for value in df[col]:
            if pd.notna(value):
                total=total+value
                count=count+1
        mean[col]=total/count if count!=0 else None

for i,j in mean.items():
    print(i," : " ,j)
```

```python
medians={}

for col in df.columns:
    if df[col].dtype in ['int','float64']:
        values=sorted([x for x in df[col]])
        n=len(values)
        if n==0:
            medians[col]=None
        elif n%2==1:
            medians[col]=values[n//2]
        else:
            medians[col]=(values[n//2-1]+values[n//2])/2

for i,j in medians.items():
    print(i," : " ,j)
```

```python
modes = {}
from collections import Counter

for col in df.columns:
    values = [x for x in df[col] if pd.notna(x)]
    if values:
        most_common = Counter(values).most_common()
        max_freq = most_common[0][1]
        mode_values = [key for key, val in most_common if val == max_freq]
        modes[col] = mode_values[0] if len(mode_values) == 1 else mode_values
    else:
        modes[col] = None

for i,j in modes.items():
    print(i," : " ,j)
```
